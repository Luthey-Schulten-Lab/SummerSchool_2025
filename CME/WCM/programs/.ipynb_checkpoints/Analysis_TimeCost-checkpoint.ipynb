```python
import re

log_filepath = '../output/log.txt'
TimeCost = {}

# TimeCost['CME Initiation'] = []
# TimeCost['CME Run'] = []
TimeCost['ODE Initiation'] = []
TimeCost['ODE Run'] = []
TimeCost['LM FileSaving'] = []
TimeCost['CMEODE run'] = []
TimeCost['CSV'] = []
hookSimulationTime = []

# float_pattern = r'[+-]?\d*\.\d+([eE][+-]?\d+)?'
# [+-]? optionally matches a leading sign (+ or -).
# \d* matches zero or more digits before the decimal point, allowing for numbers like .5.
#  \.\d+ ensures that there must be a decimal point followed by at least one digit, effectively excluding integers.
#  ([eE][+-]?\d+)? optionally matches the exponent part, allowing for scientific notation.

float_pattern = r'[+-]?\d*\.\d+'
with open(log_filepath,'r') as logfile:
    for line in logfile:
        if line.startswith('The simulation start'):
            TimeCost['StartTime'] = line
            print(TimeCost['StartTime'])
        elif line.startswith('LM file Saved'):
            savingTime = float(re.findall(float_pattern, line)[0])
            TimeCost['LM FileSaving'].append(savingTime)
        elif line.startswith('ODE object Initialized'):
            odeIniTime  = float(re.findall(float_pattern, line)[0])
            TimeCost['ODE Initiation'].append(odeIniTime)
        elif line.startswith('ODE simulation of'):
            odeRunTime = float(re.findall(float_pattern, line)[0])
            TimeCost['ODE Run'].append(odeRunTime)
        elif line.startswith('CMEODE run'):
            cmeodeRunTime = float(re.findall(float_pattern, line)[0])
            TimeCost['CMEODE run'].append(cmeodeRunTime)
        elif line.startswith('CSV File'):
            csvTime = float(re.findall(float_pattern, line)[0])
            TimeCost['CSV'].append(csvTime)
        elif line.startswith('HookSimulation'):
            hookTime = float(re.findall(float_pattern, line)[0])
            hookSimulationTime.append(hookTime)
        elif line.startswith('The simulation ends at'):
            TimeCost['EndTime'] = line
            print(TimeCost['EndTime'])
```

```python
from datetime import datetime
Start = '2024-02-22 13:32:11.773646'
End = '2024-02-23 06:26:03.777677'

Start = datetime.strptime(Start, '%Y-%m-%d %H:%M:%S.%f')
End = datetime.strptime(End, '%Y-%m-%d %H:%M:%S.%f')

TotalTime = (End - Start).total_seconds()/3600
print('Simulation Took ' + str(TotalTime) + ' Hours')

def func(pct, allvals):
    absolute = pct/100.*np.sum(allvals)
    return "{:.1f}%\n({:.2f} h)".format(pct, absolute)

odeIniCost = np.sum(TimeCost['ODE Initiation'])/3600
odeRunCost = np.sum(TimeCost['ODE Run'])/3600
LMfileSavingCost = np.sum(TimeCost['LM FileSaving'])/3600
csvCost = np.sum(TimeCost['CSV'])/3600
otherCost = TotalTime - odeIniCost - odeRunCost - LMfileSavingCost - csvCost
print(odeIniCost, odeRunCost, LMfileSavingCost, csvCost)

# Data to plot
labels = 'Initiation of ODE', 'ODE Run', 'CSV FileSaving', 'LM FileSaving', 'Others'
Costs = [odeIniCost, odeRunCost, csvCost, LMfileSavingCost, otherCost]  # Example data
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'Red']
explode = (0, 0, 0, 0, 0)  # explode the 1st slice (i.e., 'Category A')

# Plot
plt.pie(Costs, explode=explode, labels=labels, colors=colors,
autopct=lambda pct: func(pct, Costs), shadow=True, startangle=140)

plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Analyze the hookTime
import copy
hookTime = copy.deepcopy(hookSimulationTime)
# Insert multiples of 60 since hookTime is not printed at these moments in this logfile
for i in range(105):
    hookTime.insert((i+1)*60-1,(i+1)*60)

timespan = np.linspace(1,6300,6300)

import matplotlib.pyplot as plt

erro = hookTime - timespan

plt.hist(erro, bins = 100)
plt.title('Distribution of Lag Time',fontsize = 16)
plt.xlabel('Lag Time/Second',fontsize = 14)
plt.savefig('../output/Distribution of Hook Lag Time.png')
```
